```python
#Traemos solo las vairables numericas que nos interesan para generar la correlacion
variables_correlacion= ['LC', 'MAT', 'SC', 'NAT', 'ING']
df_correlacion= df[variables_correlacion]
print(df_correlacion)
```

```python
#Visualizacion de Datos Mapa de calor Correlacion Materias
plt.figure(figsize=(8,6))
sns.heatmap(df_correlacion.corr(), annot=True, cmap='coolwarm')
```

```python
#Histograma Variable Lectura Critica
plt.figure(figsize=(8,6))
sns.histplot(df_correlacion['LC'], kde=False, bins=30)
```

```python
#Histograma Variable Ciencias Sociales
plt.figure(figsize=(8,6))
sns.histplot(df_correlacion['SC'], kde=False, bins=30)
```

```python
#Histograma Variable Ciencias Naturales
plt.figure(figsize=(8,6))
sns.histplot(df_correlacion['NAT'], kde=False, bins=30)
```

```python
#Histograma Variable Ingles
plt.figure(figsize=(8,6))
sns.histplot(df_correlacion['ING'], kde=False, bins=30)
```
